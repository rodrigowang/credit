```python
import numpy as np
import pandas as pd
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
sns.set(style="darkgrid")
```

```python
dados = pd.read_csv('cs-training.csv')
```

```python
dados = dados.drop('Unnamed: 0', 1)
dados.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dados.iloc[:,1:12], dados['SeriousDlqin2yrs'], test_size=0.25, random_state=42)
```

```python
#Normalizar os dados
from sklearn import preprocessing
scalerX = preprocessing.StandardScaler().fit(X_train.dropna())
scalerY = preprocessing.StandardScaler().fit(y_train.dropna())
```

```python
def normalizarVar(dfModificado, dfOriginal):
    df1 = pd.DataFrame(index = dfModificado.index.values)
    for i in dfOriginal.columns:
        valores = (dfModificado[i] - np.mean(dfOriginal[i]))/np.std(dfOriginal[i])
        df1[i]=pd.Series(valores).values
    return df1
```

```python
X_train_scaled = normalizarVar(X_train, X_train)
X_test_scaled = normalizarVar(X_test, X_train)
y_train_scaled = (y_train - np.mean(y_train))/np.std(y_train)
y_test_scaled = (y_test - np.mean(y_train))/np.std(y_train)
```
