```python
#Distribuição das variáveis explicativas
for col in dados.columns[1:]: 
    g = sns.FacetGrid(dados, col="SeriousDlqin2yrs", margin_titles=True)
    #bins = np.linspace(0, 60, 13)
    g.map(plt.hist, col, color="steelblue", lw=0, normed = True)
```
