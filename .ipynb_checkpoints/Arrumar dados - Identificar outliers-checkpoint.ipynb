```python
def buscarOut(dados, var):
    from sklearn import linear_model
    # Robustly fit linear model with RANSAC algorithm
    model = linear_model.LinearRegression()
    X = dados.dropna().drop([var], axis = 1)
    y = dados.dropna()[var]
    model.fit(X, y)
    previsoes = model.predict(X)
    residuos = y - previsoes
    return(residuos)
```

```python
d = dict()

for var in X_train.columns:
    residuos = buscarOut(X_train,var)
    d[var] = residuos[residuos>50].index.values
```
