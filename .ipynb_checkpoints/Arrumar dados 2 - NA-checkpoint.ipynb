```python
#imputar a mediana
from sklearn.base import BaseEstimator, RegressorMixin
from numpy import median, sum, sqrt

class ImputMediana(BaseEstimator, RegressorMixin):  
    """An example of classifier"""

    def __init__(self, intValue=0, stringParam="defaultValue", otherParam=None):
        """
        Called when initializing the classifier
        """
        self.intValue = intValue
        self.stringParam = stringParam

        # THIS IS WRONG! Parameters should have same name as attributes
        self.differentParam = otherParam 


    def fit(self, X, y=None):

        self.median = median(X)  # mean + intValue

        return self


    def predict(self, X, y=None):

        return([self.median for x in X])
    
    def score(self, X, y=None):
        # counts number of values bigger than mean
        return(sqrt(sum((X - self.predict(X)) ** 2) / len(X))) 
```

```python
mediana = ImputMediana()
mediana.fit(y)
print mediana.score(y)
scoreMediana = cross_val_score(mediana, y, y, cv=5, n_jobs=1, scoring = 'neg_mean_squared_error')
```
